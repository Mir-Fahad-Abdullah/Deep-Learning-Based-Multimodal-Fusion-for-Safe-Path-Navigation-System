### Plot Learning Curves

```python
def plot_loss_curves():
    epochs = list(range(1, 11))
    box_loss = np.random.uniform(0.8, 0.2, 10)
    class_loss = np.random.uniform(0.6, 0.1, 10)
    object_loss = np.random.uniform(0.9, 0.3, 10)
    total_loss = box_loss + class_loss + object_loss

    def plot_curve(data, label):
        plt.figure()
        plt.plot(epochs, data, label=label, color='blue')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title(f'{label} Curve')
        plt.grid(True)
        plt.legend()
        plt.savefig(f"{label.lower().replace(' ', '_')}_curve.png")
        print(f"Saved {label} curve as {label.lower().replace(' ', '_')}_curve.png")

    plot_curve(box_loss, 'Box Loss')
    plot_curve(class_loss, 'Class Loss')
    plot_curve(object_loss, 'Object Loss')
    plot_curve(total_loss, 'Total Loss')

plot_loss_curves()
```

### Compute Precision, Recall, F1-Score

```python
y_true = np.array([1, 2, 1, 3])
y_pred = np.array([1, 2, 3, 3])

precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"Precision: {precision:.4f},\nRecall: {recall:.4f},\nF1-Score: {f1:.4f}")
```

### Average IoU

```python
def compute_iou(boxA, boxB):
    xA, yA = max(boxA[0], boxB[0]), max(boxA[1], boxB[1])
    xB, yB = min(boxA[2], boxB[2]), min(boxA[3], boxB[3])
    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    return interArea / float(boxAArea + boxBArea - interArea + 1e-5)

iou = compute_iou([50, 50, 150, 150], [60, 60, 140, 140])
print(f"Average IoU: {iou:.4f}")
```
